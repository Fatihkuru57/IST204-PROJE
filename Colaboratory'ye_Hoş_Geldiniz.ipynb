```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
quantiles = df1['İşsizlikOranı'].quantile([0.25, 0.5, 0.75])
print(quantiles)
```

```
df1_total = df1[df1['Cinsiyet'] == 'Toplam']
years = df1_total['yıllar']
workforce = df1_total['İşgücü']
employment = df1_total['İstihdam']
unemployment = df1_total['İşsiz']
plt.figure(figsize=(12, 6))
plt.plot(years, workforce, label='İşgücü', color='blue', marker='o')
plt.plot(years, employment, label='İstihdam', color='green', marker='o')
plt.plot(years, unemployment, label='İşsiz', color='red', marker='o')
plt.xlabel('Yıllar')
plt.ylabel('Kişi Sayısı')
plt.title('15 - 24 Yaş Arasındaki Genç Nüfusun İşgücü, İstihdam ve İşsizlik Verileri')
plt.legend()
plt.grid(True)
plt.show()
```

```
df1_total = df1[df1['Cinsiyet'] == 'Toplam']
x = df1_total['yıllar']
y_workforce = df1_total['İşgücü']
y_employment = df1_total['İstihdam']
y_unemployment = df1_total['İşsiz']
dx = np.diff(x)
dy_workforce = np.diff(y_workforce)
dy_employment = np.diff(y_employment)
dy_unemployment = np.diff(y_unemployment)
X = x[:-1]
Y_workforce = y_workforce[:-1]
Y_employment = y_employment[:-1]
Y_unemployment = y_unemployment[:-1]
plt.figure(figsize=(12, 6))
plt.quiver(X, Y_workforce, dx, dy_workforce, angles='xy', scale_units='xy', scale=1, color='blue', label='İşgücü')
plt.quiver(X, Y_employment, dx, dy_employment, angles='xy', scale_units='xy', scale=1, color='green', label='İstihdam')
plt.quiver(X, Y_unemployment, dx, dy_unemployment, angles='xy', scale_units='xy', scale=1, color='red', label='İşsiz')
plt.xlabel('Yıllar')
plt.ylabel('Değerler')
plt.title('15 - 24 Yaş Arasındaki Genç Nüfusun İşgücü, İstihdam ve İşsizlik Değişimleri (Quiver Grafiği)')
plt.legend()
plt.grid(True)
plt.show()
```

```
fig, axs = plt.subplots(3, 1, figsize=(15, 15), sharex=True)
axs[0].plot(df1['yıllar'], df1['İşgücü'], marker='o', linestyle='-', color='b')
axs[0].set_title('İşgücü')
axs[0].set_ylabel('İşgücü (bin)')
axs[1].plot(df1['yıllar'], df1['İstihdam'], marker='s', linestyle='-', color='g')
axs[1].set_title('İstihdam')
axs[1].set_ylabel('İstihdam (bin)')
axs[2].plot(df1['yıllar'], df1['İşsiz'], marker='d', linestyle='-', color='r')
axs[2].set_title('İşsiz')
axs[2].set_xlabel('Yıllar')
axs[2].set_ylabel('İşsiz (bin)')
plt.tight_layout()
plt.show()
```

```
years = df1['Yıllar'].unique()
months = df1['Aylar'].unique()
n_years = len(years)
n_months = len(months)
year_indices = {year: i for i, year in enumerate(years)}
month_indices = {month: i for i, month in enumerate(months)}
grid = np.zeros((n_years, n_months))
for index, row in df1.iterrows():
    year = row['Yıllar']
    month = row['Aylar']
    population = row['İstihdam']
    grid[year_indices[year], month_indices[month]] = population
plt.figure(figsize=(12, 8))
plt.pcolormesh(grid, cmap='Blues')
plt.colorbar(label='İstihdam')
plt.xticks(np.arange(0.5, n_months, 1), months, rotation=90)
plt.yticks(np.arange(0.5, n_years, 1), years)
plt.xlabel('Aylar')
plt.ylabel('Yıllar')
plt.title('İstihdamın  Aylara Ve Yıllara Göre Dağılımı')
plt.tight_layout()
plt.show()
```

```
years = df1['Yıllar']
population = df1['İşsiz']
error = np.random.uniform(low=500, high=1000, size=len(population))
plt.figure(figsize=(10, 6))
plt.errorbar(years, population, yerr=error, fmt='o', color='blue', ecolor='red', elinewidth=2, capsize=4)
plt.xlabel('Yıl')
plt.ylabel('İşsizlik')
plt.title('Yıllara Göre İşsizlik')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
